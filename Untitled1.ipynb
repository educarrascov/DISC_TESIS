```python
from matplotlib import pyplot as plt
import matplotlib.patches as patches

# Create the figure and axis
fig, ax = plt.subplots(figsize=(12, 8))

# Define positions and labels
boxes = [
    {"text": "1. Problem Definition and Input Modeling", "xy": (0.5, 0.9)},
    {"text": "2. Mathematical Formulation of TTAP", "xy": (0.5, 0.72)},
    {"text": "3. Resolution Strategies", "xy": (0.5, 0.54)},
    {"text": "4. Experimental Design", "xy": (0.5, 0.36)},
    {"text": "5. Evaluation and Analysis", "xy": (0.5, 0.18)},
]

# Draw boxes and arrows
for i, box in enumerate(boxes):
    ax.add_patch(patches.FancyBboxPatch(
        (box["xy"][0]-0.4, box["xy"][1]-0.06), 0.8, 0.1,
        boxstyle="round,pad=0.02", edgecolor="black", facecolor="lightblue", linewidth=1.5
    ))
    ax.text(box["xy"][0], box["xy"][1], box["text"],
            ha='center', va='center', fontsize=11, fontweight='bold')
    if i < len(boxes) - 1:
        ax.annotate('', xy=(0.5, box["xy"][1] - 0.1), xytext=(0.5, box["xy"][1] - 0.17),
                    arrowprops=dict(arrowstyle='->', lw=2))

# Adjust and clean up plot
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')
plt.tight_layout()
plt.show()
```
