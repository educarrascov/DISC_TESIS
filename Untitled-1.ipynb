```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.interpolate import make_interp_spline

# Datos
categories = ["Exact", "Heuristic", "Metaheuristic", "Machine Learning"]
values = np.array([6, 9, 9, 8])

# Repetir primer punto al final para cerrar el grfico radial
values_closed = np.append(values, values[0])

# ngulos base
angles = np.linspace(0, 2 * np.pi, len(values_closed))

# Crear interpolacin suave (spline)
angles_smooth = np.linspace(0, 2 * np.pi, 300)
spline = make_interp_spline(angles, values_closed, k=3)
values_smooth = spline(angles_smooth)

# Estilo seaborn
sns.set_theme(style="whitegrid")

# Plot radial
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)

ax.plot(angles_smooth, values_smooth, linewidth=3)
ax.fill(angles_smooth, values_smooth, alpha=0.3)

# Etiquetas de los ejes
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, fontsize=12)

# Ttulo
plt.title("WTA Solution Methods by Category (Radial Smooth Chart)", fontsize=14)

plt.show()
```

```python
exact_years = [1971, 1973, 1990, 2003, 2007, 2008, 2020]
heuristic_years = [1991, 2008, 2012, 2013, 2016, 2018, 2020, 2021, 2022]
meta_years = [1997, 2008, 2009, 2010, 2013, 2015, 2016, 2017, 2018, 2021, 2022]
ml_years = [2008, 2009, 2012, 2021, 2022, 2023]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data
data = {
    "Exact": exact_years,
    "Heuristic": heuristic_years,
    "Metaheuristic": meta_years,
    "ML": ml_years
}

# Transform into rows
rows = []
for cat, years in data.items():
    for y in years:
        rows.append([cat, y])

df = pd.DataFrame(rows, columns=["Category", "Year"])

# Count per year-category
counts = df.groupby(["Year", "Category"]).size().reset_index(name="Count")

# Plot
sns.set_theme(style="whitegrid")
plt.figure(figsize=(14, 6))
sns.barplot(
    data=counts,
    x="Year",
    y="Count",
    hue="Category"
)

plt.title("Timeline of WTA Methods by Category (Year-Level)", fontsize=16)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline

# =======================
# 1) DATA REAL (sin 1970s)
# =======================
exact_years      = [1990, 2003, 2007, 2008, 2020]
heuristic_years  = [1991, 2008, 2012, 2013, 2016, 2018, 2020, 2021, 2022]
meta_years       = [1997, 2008, 2009, 2010, 2013, 2015, 2016, 2017, 2018, 2021, 2022]
ml_years         = [2008, 2009, 2012, 2021, 2022, 2023]

data = {
    "Exact": exact_years,
    "Heuristic": heuristic_years,
    "Metaheuristic": meta_years,
    "Machine Learning": ml_years
}

# =======================
# 2) BUILD DATAFRAME
# =======================
rows = []
for cat, years in data.items():
    for y in years:
        decade = (y // 10) * 10
        rows.append([cat, y, decade])

df = pd.DataFrame(rows, columns=["Category", "Year", "Decade"])

# FILTRAR SOLO DESDE 1990
df = df[df["Decade"] >= 1990]

# =======================
# 3) MATRIX (decade x category)
# =======================
decades = sorted(df["Decade"].unique())      # 1990, 2000, 2010, 2020
categories = ["Exact", "Heuristic", "Metaheuristic", "Machine Learning"]

matrix = np.zeros((len(decades), len(categories)))

for i, dec in enumerate(decades):
    for j, cat in enumerate(categories):
        matrix[i, j] = len(df[(df["Decade"] == dec) & (df["Category"] == cat)])

# =======================
# 4) RADAR PLOT
# =======================
N = len(categories)
angles = np.linspace(0, 2 * np.pi, N, endpoint=False)

sns.set_theme(style="whitegrid")
plt.figure(figsize=(9,9))
ax = plt.subplot(111, polar=True)

palette = sns.color_palette("tab10", n_colors=len(decades))

# Curvas = dcadas
for idx, decade in enumerate(decades):
    vals = matrix[idx]
    
    vals_ext = np.append(vals, vals[0])        # cerrar curva
    ang_ext  = np.append(angles, 2*np.pi)      # cerrar ngulo
    
    theta_dense = np.linspace(0, 2*np.pi, 400)
    spline = make_interp_spline(ang_ext, vals_ext, k=3)
    smooth_vals = spline(theta_dense)

    ax.plot(theta_dense, smooth_vals, linewidth=2.5, label=str(decade), color=palette[idx])
    ax.fill(theta_dense, smooth_vals, alpha=0.18, color=palette[idx])

# Etiquetas de categoras
ax.set_xticks(angles)
ax.set_xticklabels(categories, fontsize=12)

ax.set_title("Papers analizados con mtodos de solucin de WTA por Categoria \n(19902020)", 
             fontsize=16, pad=20)

plt.legend(loc="upper right", bbox_to_anchor=(1.2, 1.15), title="Decade")
plt.tight_layout()
plt.show()
```
