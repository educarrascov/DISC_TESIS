```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.interpolate import make_interp_spline

# Datos
categories = ["Exact", "Heuristic", "Metaheuristic", "Machine Learning"]
values = np.array([6, 9, 9, 8])

# Repetir primer punto al final para cerrar el grfico radial
values_closed = np.append(values, values[0])

# ngulos base
angles = np.linspace(0, 2 * np.pi, len(values_closed))

# Crear interpolacin suave (spline)
angles_smooth = np.linspace(0, 2 * np.pi, 300)
spline = make_interp_spline(angles, values_closed, k=3)
values_smooth = spline(angles_smooth)

# Estilo seaborn
sns.set_theme(style="whitegrid")

# Plot radial
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)

ax.plot(angles_smooth, values_smooth, linewidth=3)
ax.fill(angles_smooth, values_smooth, alpha=0.3)

# Etiquetas de los ejes
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, fontsize=12)

# Ttulo
plt.title("WTA Solution Methods by Category (Radial Smooth Chart)", fontsize=14)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ===============================
# 1) DATOS
# ===============================
# Categoras (ngulo)
categories = ["Exact", "Heuristic", "Metaheuristic", "Machine Learning"]

# Aos (radio)
years = [2021, 2022, 2023, 2024]

# Matriz de valores: filas = aos, columnas = categoras
# EJEMPLO (reemplaza estos nmeros por los tuyos)
values = np.array([
    [4,  6,  5,  3],   # 2021
    [5,  7,  6,  4],   # 2022
    [6,  8,  8,  6],   # 2023
    [6,  9,  9,  8],   # 2024
])

# Comprobacin rpida
assert values.shape == (len(years), len(categories)), "Dimensiones de 'values' no coinciden"


# ===============================
# 2) CONFIGURACIN POLAR
# ===============================
n_cat = len(categories)
n_years = len(years)

# Bordes angulares (uno ms para cerrar el crculo)
theta_edges = np.linspace(0, 2 * np.pi, n_cat + 1)

# Bordes radiales: un anillo por ao
# Por ejemplo, anillo 1 = year[0], anillo 2 = year[1], etc.
r_edges = np.arange(0, n_years + 1, 1)

# Para pcolormesh, necesitamos una cuadrcula 2D de bordes
Theta, R = np.meshgrid(theta_edges, r_edges)

# ===============================
# 3) ESTILO SEABORN
# ===============================
sns.set_theme(style="white")

# Usamos un colormap de seaborn (puedes cambiar "viridis" por "mako", "rocket", etc.)
cmap = sns.color_palette("viridis", as_cmap=True)

# ===============================
# 4) FIGURA POLAR
# ===============================
fig = plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)

# pcolormesh espera una matriz de valores con shape (n_years, n_cat)
pcm = ax.pcolormesh(Theta, R, values, cmap=cmap, shading="auto")

# ===============================
# 5) ETIQUETAS Y FORMATO
# ===============================

# ngulos centrales para etiquetas de categoras
theta_centers = (theta_edges[:-1] + theta_edges[1:]) / 2
ax.set_xticks(theta_centers)
ax.set_xticklabels(categories, fontsize=12)

# Radios centrales para etiquetar los aos
r_centers = np.arange(0.5, n_years + 0.5, 1)
ax.set_yticks(r_centers)
ax.set_yticklabels(years, fontsize=11)

# Opcional: quitar etiquetas radiales si quieres algo ms limpio
# ax.set_yticklabels([])

# Ttulo
ax.set_title("WTA Solution Methods by Category and Year\n(Polar Radial Heatmap)", fontsize=14, pad=20)

# Barra de color
cbar = plt.colorbar(pcm, pad=0.1, shrink=0.8)
cbar.set_label("Valor (n de mtodos / papers / etc.)", fontsize=11)

# Ajustar layout
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.interpolate import make_interp_spline

# Data
categories = ["Exact", "Heuristic", "Metaheuristic", "Machine Learning"]
values = np.array([6, 9, 9, 8])

# Close the circle
values_closed = np.append(values, values[0])
angles = np.linspace(0, 2 * np.pi, len(values_closed))

# Smooth curve
theta_smooth = np.linspace(0, 2*np.pi, 300)
spline = make_interp_spline(angles, values_closed, k=3)
values_smooth = spline(theta_smooth)

# Plot aesthetics
sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)

ax.plot(theta_smooth, values_smooth, linewidth=3)
ax.fill(theta_smooth, values_smooth, alpha=0.25)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, fontsize=13)
ax.set_yticklabels([])

plt.title("WTA Solution Methods by Category", fontsize=16, pad=20)
plt.show()
```

```python
exact_years = [1971, 1973, 1990, 2003, 2007, 2008, 2020]
heuristic_years = [1991, 2008, 2012, 2013, 2016, 2018, 2020, 2021, 2022]
meta_years = [1997, 2008, 2009, 2010, 2013, 2015, 2016, 2017, 2018, 2021, 2022]
ml_years = [2008, 2009, 2012, 2021, 2022, 2023]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data
data = {
    "Exact": exact_years,
    "Heuristic": heuristic_years,
    "Metaheuristic": meta_years,
    "ML": ml_years
}

# Transform into rows
rows = []
for cat, years in data.items():
    for y in years:
        rows.append([cat, y])

df = pd.DataFrame(rows, columns=["Category", "Year"])

# Count per year-category
counts = df.groupby(["Year", "Category"]).size().reset_index(name="Count")

# Plot
sns.set_theme(style="whitegrid")
plt.figure(figsize=(14, 6))
sns.barplot(
    data=counts,
    x="Year",
    y="Count",
    hue="Category"
)

plt.title("Timeline of WTA Methods by Category (Year-Level)", fontsize=16)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Add decade column
df["Decade"] = (df["Year"] // 10) * 10

# Count per decade-category
decade_counts = df.groupby(["Decade", "Category"]).size().reset_index(name="Count")

plt.figure(figsize=(12, 6))
sns.barplot(
    data=decade_counts,
    x="Decade",
    y="Count",
    hue="Category"
)

plt.title("WTA Methods by Category and Decade", fontsize=16)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.interpolate import make_interp_spline  # pip install scipy

# =========================================
# 1) DATOS (CAMBIA ESTO POR LOS TUYOS)
# =========================================
# Aos
years = np.array([2018, 2019, 2020, 2021, 2022, 2023])

# Diccionario: categora -> valores por ao
# EJEMPLO (reemplaza estos valores por tus datos reales)
data = {
    "Exact":          np.array([2, 3, 4, 5, 5, 6]),
    "Heuristic":      np.array([3, 4, 5, 6, 7, 9]),
    "Metaheuristic":  np.array([2, 3, 4, 5, 7, 9]),
    "Machine Learning": np.array([1, 2, 3, 4, 6, 8]),
}

# Comprobacin rpida
n_years = len(years)
for k, v in data.items():
    assert len(v) == n_years, f"La categora {k} no tiene un valor por cada ao"


# =========================================
# 2) MAPEO A COORDENADAS POLARES
# =========================================
# Normalizar aos al rango [0, 2]
years_norm = (years - years.min()) / (years.max() - years.min())
theta = years_norm * 2 * np.pi

# Estilo general
sns.set_theme(style="whitegrid")
palette = sns.color_palette("tab10", n_colors=len(data))

# =========================================
# 3) FIGURA POLAR
# =========================================
fig = plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)

for (cat, vals), color in zip(data.items(), palette):
    # --- Suavizado opcional con spline ---
    # Aadimos el primer punto al final para "cerrar" la curva
    theta_closed = np.append(theta, theta[0])
    vals_closed = np.append(vals, vals[0])

    # Interpolacin suave
    theta_smooth = np.linspace(theta_closed.min(), theta_closed.max(), 300)
    spline = make_interp_spline(theta_closed, vals_closed, k=3)
    vals_smooth = spline(theta_smooth)

    # Curva suave
    ax.plot(theta_smooth, vals_smooth, label=cat, color=color, linewidth=2)

    # Puntos originales (para que se vea dnde estn los datos reales)
    ax.scatter(theta, vals, color=color, s=30, zorder=3)

# =========================================
# 4) ETIQUETAS Y FORMATO
# =========================================

# Etiquetas de aos en el ngulo
ax.set_xticks(theta)
ax.set_xticklabels(years, fontsize=11)

# Etiquetas del radio (puedes ajustar)
max_val = max(v.max() for v in data.values())
ax.set_ylim(0, max_val * 1.1)

ax.set_yticks(range(0, int(max_val) + 1))
ax.set_yticklabels(range(0, int(max_val) + 1), fontsize=10)

# Ttulo y leyenda
ax.set_title("Evolucin por Ao y Categora\n(Radial Timeline)", fontsize=14, pad=20)
ax.legend(loc="upper right", bbox_to_anchor=(1.3, 1.1))

plt.tight_layout()
plt.show()
```
